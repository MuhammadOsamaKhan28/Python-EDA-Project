```python
# dropping null values
data.dropna(inplace=True)
```

**EDA**
Data Analysis

```python
sns.histplot(data['price'])
```

```python
# sns.boxplot(data['price'])
df = data[data['price'] < 1500]

sns.boxplot(data=df,x='price')
```

```python
plt.figure(figsize=(8,5))

sns.histplot(data=df,x='price',bins=100,color='green')
plt.ylabel("Frequency")
plt.xlabel("Price")
plt.title("Price Distribution")
```

```python
df.columns
```

```python
plt.figure(figsize=(8,5))

sns.histplot(data=df,x='availability_365',bins=100,color='blue')
plt.ylabel("Frequency")
plt.xlabel("availability_365")
plt.title("availability_365 Distribution")
```

```python
df.groupby(by='neighbourhood_group')['price'].mean()
```

```python
df['Price per bedroom'] = df['price']/df['beds']
```

```python
df.groupby(by='neighbourhood_group')['Price per bedroom'].mean()
```

**Bi Variable Analysis**


```python
df.columns
```

```python
sns.barplot(data = df,x='neighbourhood_group',y='price',hue='room_type')
```

```python
plt.figure(figsize=(8,5))
plt.title('Price Distribution')
sns.scatterplot(data=df,x='number_of_reviews',y='price',hue='neighbourhood_group')
```

```python
sns.pairplot(df,vars=['price','minimum_nights','number_of_reviews','availability_365'],hue='room_type')
```

```python
plt.figure(figsize=(10,7))
sns.scatterplot(data=df,x=
                'longitude',y='latitude',hue='room_type')
plt.title('Price Distribution')
```

```python
#heatmap
corr = df[['latitude','longitude','price','minimum_nights','availability_365','number_of_reviews','beds']].corr()
corr
```

```python
plt.title('Correlation Heatmap')
sns.heatmap(corr,annot=True,cmap='coolwarm')
```
